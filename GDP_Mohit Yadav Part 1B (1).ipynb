```python
#considering all data of item and 2014-15 of all states

data=[]
for location in all_files:
    df=pd.read_csv(location,index_col=None,header=0,encoding='latin1')
    df=df[['Item','2014-15']]
    df=df.set_index('Item')
    df=df.T
    df.loc[:,'State']= location.split('\\')[-1].split('.')[0]
    data.append(df)
    
    
```

### Why is (Primary+ Secondary+Tertiary) not equal to total GDP ?
#### Primary + Secondary + Tertiary = TOTAL GSVA at basic prices
#### TOTAL GSVA at basic prices + Taxes on Products - Subsidies on products = Gross State Domestic Product

```python
#Categorise the states into four groups based on the GDP per capita(C1,C2, C3, C4)
#where C1 would have highest per capita GDP
#C4 will have the lowest
#adding new cateory column for the same 

category=[]

for row in data2['Per Capita GSDP (Rs.)']:
    if row > 153246 and row <=271793:
        category.append('C1')
    elif row > 104977 and row <=153246:
        category.append('C2')
    elif row >65187 and row <=104977:
        category.append('C3')
    else:
        category.append('C4')
        
        
data2['Category']=category

data2
        
        
        
        
        
        
```
